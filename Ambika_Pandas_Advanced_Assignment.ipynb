## Groupby

```python
# Create a dataframe using dictionary of your choice

data =pd.DataFrame({'Name':['Ankit','Sumit','Shreya','Sam','Joe','Rishi','Arna'],'Age':[20,21,22,21,23,24,22],'Univ':['BHU','JNU','DU','DU','UoR','UoR','BHU'],'YOE':[1,2,3,2,1,3,1]})

data
```

```python
# Use Groupby of single column with aggregate sum()
df_grouped=data.groupby('Univ')
BHU=df_grouped.get_group('BHU')
BHU['Age'].sum()
```

```python
# Use Groupby of single column with aggregate count()

BHU['Age'].count()
```

```python
# Use Groupby of single column with aggregate min() and max()

print('Minimum age: ', BHU['Age'].min())
print('\n Maximum age: ',BHU['Age'].max())
```

```python
# Use Groupby of any 2 columns with aggregate mean()

df_grouped2=data.groupby(['Univ','YOE'])
BHU=df_grouped2.get_group(('BHU',1))
BHU['Age'].mean()
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()
print('Minimum age: ', BHU['Age'].min())
print('\n Maximum age: ',BHU['Age'].max())

```
